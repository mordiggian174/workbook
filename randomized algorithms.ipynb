**Рандомизированные алгоритмы**

Самый простой рандомизированный алгоритм - поиск наименьшего размера в графе через стягивание случайного ребра.

Чтобы вероятность ошибки сделать сколь угодно маленькой, прибегают к многочисленным параллельным запускам алгоритма.

Различают два вида рандомизированных алгоритмов (но они превращаются друг в друга):
* Лас-Вегас (время работы может быть бесконечным, как одна из стратегий в казино)
* Монте-Карло (время ограничено, но мы можем ошибиться)

Одним из классических примеров рандомизированных алгоритмов является хэш-таблица. У нее есть минусы, что для хранения $m$ элементов понадобится порядка $64m$ битов (в списочном варианте со ссылками). А можно разрешить ложно-положительные ответы с маленькой вероятностью, но сэкономить ра замере.

**Фильтр Блюма** - мы имеем $k$ хэш-функций, которые по объекту из $S$ получают свои индексы и инкрементируют соответствующий элемент массива. Важным нюансом является то, что теперь мы говорим "элемент, возможно, лежит в нашем хранилище", когда по каждому из индексов непусто. Но теперь пустоты элементов массива - это зависимые величины и их анализировать не так просто. Правильно подбирая $k$ через оптимизацию функционала правдоподобия, мы получаем экспоненциальное убывание вероятности ошибки по параметру $\frac{n}{m}$. Важной идеей в анализе было приближение независимыми случайными величинами из Пуассоновского распределения.

Также интересное применение для поиска подстрок в строке через фильтр Блюма и полиномиально-скользящие хэш-функции (алгоритм **Карпа-Рабина**). Преимущество в том, что их можно очень быстро вычислять, а вероятность ошибки достаточно мала из-за техники Блюма и подбора параметров.

Замечательно, что алгоритмы Лас-Вегас можно рассматривать как семейство детерменированных алгоритмов, среди которых случайным образом выбирается какой-то при произвольной входе $х$. Тогда можно это свести к соревнованию с двумя игроками: один подбирает распределение входов $x$, второй алгоритмов $A$.
