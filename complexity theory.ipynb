**Complexity Theory**

Хотим построить подход оценки сложности *задачи* в рамках некоторой вычислительной модели. Базируемся на *машине Тьюринга* и *задачах распознавания*.

Определяя классы сложности задач распознавания в соответствии со временем и объемом памяти, несложно показать следующее утверждение.


Таким же трюком с переходом к другому алфавиту можно показать, что 
* DSPACE($f$) = DSPACE( $\mathcal{O}(f)$ )
* DTIME($f$) = DTIME( $\mathcal{O}(f)+n$ ) (мы обязаны хотя бы прочитать вход в общем случае)
* DTIME( $\mathcal{O}(n))$ = DTIME($2n$) > DTIME ($n$) ($n$ операций только на прочтение, а далее сколь угодно можно сжимать через алфавит)

Чтобы доказывать наличие какой-то иерархии в классах сложности, часто прибегают к методу **диагонализации** - это когда в рамках ограничений класса А строят машину, которая ведет себя отлично от любой машины класса Б (там важно иметь ограничения через конструирумые во времени и пространстве функции!). Все по итогу сводится к моделированию (и инвертированию ответа) через универсальюную машину Тьюринга, которая разрастается по времени в $\log n$ раз.

* DTIME $f(n)$ < DTIME $\Omega \left(f(n)\log n\right)$
* DSPACE $f(n)$ < DSPACE $\Omega f(n)$

Чтобы ввести понятие эффективно решаемой задачи, мы замыкаем линейные (их точно не улучшить) алгоритмы через композицию и получаем класс $P$, который, как мы и хотели, не зависит от модели вычислений.

Класс $NP$ можно ввести как $P$ на недетерменированной машине Тьюринга, либо как задачу с полномиальным верификатором и сертификатом полиномиальной длины. $P \le NP \le PSPACE$ (полный перебор сертификатов). Через НМТ вводятся и классы $NL, NP, NPSPACE, NEXP$.

Перебором конфигураций и поиском в глубину можно получить следующее утверждение.
* $NL \le P$
* $NSPACE = PSPACE$

Также выделяют классы сложности, состоящие из языков-дополнений к языкам каких-то ранее определенных классов.

Для простоты анализа выделяют **полные**, **трудные задачи** и **сводимости во времени или пространстве** (для транзитивности отношения и замкнутости класса сложности). Также есть некоторые результаты по поводу того, что по решению задачи распознавания можно строить и само решение-ответ. Это объясняет то, что мы обходились только задачами распознавания языков.

В **задачах с оракулами** машинам Тьюринга разрешается строить запрос к некоторой процедуре, дающей ответ за $\mathcal{O}(1)$
Также из игр с двумя игроками и их обобщениями можно построить полные задачи в полиномиальной иерархии.

Есть ещё другая вычислительная модель, основывающаяся на булевых схемах. Там мерой сложности задачи будет размер семейства схем $\mathcal{C}_n$. Интересен факт, что эффективной конструируемости не требуется, из-за чего появляется разрыв в вычислительных мощностях моделей: задача остановки HALT разрешима на схемах. 
Практический интерес представляют задачи, для которых есть схемы полиномиального размера, это класс $P_{/poly} > P$. Такая запись не случайна: МТ, которые имеют доступ к советам (одно слово некоторой длины для каждого n) становятся эквивалентны моделям булевых схем.
