**Подход через параметризованные алгоритмы**

Для большинства сложных задач неизвестны полиномиальные алгоритмы. На практике их решают с помощью эвристик или приближенными методами. Но можно выбирать другой подход: вычленять сущность, которая описывает структурную сложность задачи, и строить алгоритм в зависимости от нее. Так можно производить редукцию данных до сравнительно малых размеров, либо адаптировать специальным образом процедуры.
